```python
def cont_disc(df):
    continuous = pd.DataFrame()
    discrete = pd.DataFrame()
    for col in df:
        if df[col].nunique() > 15:
            continuous[col] = df[col]
        elif df[col].nunique() <= 15:
            discrete[col] = df[col]
    return [continuous, discrete]
```

```python
continuous=cont_disc(customer_df)[0]
discrete=cont_disc(customer_df)[1]
```

```python
def needs_cleaning(df):
    to_clean = pd.DataFrame()
    cleaned = pd.DataFrame()
    for col in df:
        if df[col].nunique() > 5:
            to_clean[col] = df[col]
        elif df[col].nunique() <= 5:
            cleaned[col] = df[col]
    return [to_clean, cleaned]
```
